```python
import pandas as pd

eia = pd.read_csv("data_clean/eia_clean.csv")
fema = pd.read_csv("data_clean/fema_clean.csv")
noaa = pd.read_csv("data_clean/noaa_clean.csv")
```

```python
eia_state_year = (
    eia
    .groupby(['State', 'Year'], as_index=False)
    .agg(
        Total_Revenues=('Total Revenues', 'sum'),
        Total_Sales=('Total Sales', 'sum'),
        Total_Customers=('Total Customers', 'sum')
    )
)
```

```python
eia_state_year['Revenue_per_Customer'] =(
    eia_state_year['Total_Revenues'] / eia_state_year['Total_Customers']
)

eia_state_year['Sales_per_Customer'] = (
    eia_state_year['Total_Sales'] / eia_state_year['Total_Customers']
)
```

```python
fema['Year'] = pd.to_datetime(
    fema['declarationDate']
).dt.year

fema_state_year = (
    fema
    .groupby(['state', 'Year'], as_index=False)
    .agg(
        Disaster_Count=('disasterNumber', 'nunique')
    )
)

fema_state_year.rename(columns={'state': 'State'}, inplace=True)
```

```python

noaa['YEAR'] = noaa['YEAR'].astype(int)
```

```python
import pandas as pd
#cleaning

def clean_damage(x):
    if pd.isna(x):
        return 0
    x = str(x).upper().replace('$','').strip()  
    if x.endswith('K'):
        return float(x[:-1]) * 1_000
    if x.endswith('M'):
        return float(x[:-1]) * 1_000_000
    if x.endswith('B'):
        return float(x[:-1]) * 1_000_000_000
    try:
        return float(x)
    except:
        return 0


noaa['DAMAGE_PROPERTY'] = noaa['DAMAGE_PROPERTY'].apply(clean_damage)
noaa['DAMAGE_CROPS'] = noaa['DAMAGE_CROPS'].apply(clean_damage)


print(noaa[['DAMAGE_PROPERTY']].head(20))
```

```python
noaa['Casualties'] = (
    noaa['INJURIES_DIRECT'] +
    noaa['DEATHS_DIRECT']
)
```

```python
print(noaa.columns)
```

```python
noaa_state_year = (
    noaa
    .groupby(['STATE', 'YEAR'], as_index=False)
    .agg(
        Extreme_Event_Count=('EVENT_ID', 'nunique'),
        Total_Property_Damage=('DAMAGE_PROPERTY', 'sum'),
        Total_Crop_Damage=('DAMAGE_CROPS', 'sum'),
        Total_Casualties=('Casualties', 'sum')
    )
)

noaa_state_year.head(20)
```

```python
noaa_state_year['State'] = (
    noaa_state_year['State']
    .str.strip()
    .str.upper()
    .map(state_abbrev)
)
```

```python
noaa_state_year = noaa_state_year.dropna(subset=['State'])
```

```python
print(noaa_state_year[['State', 'Year']].head(10))
```

```python
final_dataset = eia_state_year.copy()

final_dataset = final_dataset.merge(
    fema_state_year,
    on=['State', 'Year'],
    how='left'
)

final_dataset = final_dataset.merge(
    noaa_state_year,
    on=['State', 'Year'],
    how='left'
)

final_dataset.head(20)
```

```python
final_dataset[['State', 'Year', 'Total_Property_Damage', 'Total_Crop_Damage']].head(15)
```

```python
numeric_cols = [
    'Disaster_Count',
    'Extreme_Event_Count',
    'Total_Property_Damage',
    'Total_Crop_Damage',
    'Total_Casualties'
]

final_dataset[numeric_cols] = final_dataset[numeric_cols].apply(pd.to_numeric, errors='coerce')

final_dataset[numeric_cols] = final_dataset[numeric_cols].fillna(0)

final_dataset.head(20)
```
